## Visualicemos los datos:


1.   `df.head(10)`: Imprime solo las 10 primeras filas del Dataframe.
2.   `df.columns`: Nombre de las columnas.
3.   `df['col_1']`: Selecciono solo la columna llamada "col_1".
4.   `df.info()`: Tipo de datos de las columnas(int,float,object) junto con otros datos interesantes.
5.   `df.isna().sum()`: Cantidad de valores faltantes por columna.
6.   `df['col_1'].fillna(0)`: Relleno los datos faltantes de mi columna 'col_1' con el valor 0.
7.   `df.duplicated().sum()`: Cantidad de valores duplicados.
8.   `df.drop_duplicated()`: Elimina filas duplicadas.
9.   `df.loc[]`: Selecciona columnas y filas con sus nombres.
10.   `df.iloc[]`: Selecciona columnas y filas con índices.
11.   `df['col_1'].nunique()`: cuantos valores unicos poseo en mi columna "col_1".

**Adicionalmente:**

1.   `df[df['genero']=='Otro']`: Filtramos nuestro dataframe solo las filas que contengan la columna llamada "genero" el valor "Otro'.
2.   `df[ (df['genero']=='Otro') & (df['edad']>50) ]`: Filtramos nuestro dataframe solo las filas que contengan la columna llamada "genero" el valor "Otro' y además en la columna "edad" contengan el valor 50
