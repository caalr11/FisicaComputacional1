$x_{n+1}=\frac{1}{2}(x_n+\frac{S}{x_n})$


```
import numpy as np

print("Este programa sirve para aproximar un numero con su raiz por el metodo Babiliano.")
print("Ingrese el numero al cual querra aproximar su raiz:")

S=float(input())
Sr=np.sqrt(S) #La raiz exacta

print("Ingrese valor inicial de x0:")

x0=float(input())

i=1
F=1/2*(x0+(S/x0))

while abs(S-F) > 0.01:
  print("Intento", i, F)
  F=1/2*(F+(S/F))
  i=i+1
  if i >= 10000: #Esto evitara que el loop corra infinitamente.
    break

print("El valor aproximado de la raiz de ", S, "es:", F , ". Con error de:", abs(np.sqrt(S)-F))
```
