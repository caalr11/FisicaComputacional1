**Ejercicio 1, Area de un circulo**
$A=\pi r^2$



```
#Area de un circulo

print("Calculacion del area de un circulo, con radio r")
def ac(x):
  return np.pi*x**2  #formula que se utilizara

print("Ingrese valor de radio")
r=float(input())  

ac=ac(r)

print("Area del circulo con radio r:", ac)
print("Area del circulo con radio r y dos decimales ={:.2f}".format(ac))
```

```
#Area de una elpise
print("Calculo del area de una elipse.")
def ae(x,y):
  return x*y*np.pi #Definicion de formula.

print("Ingrese el valor del semieje mayor de la elipse y despues el valor del semieje menor:")
a=float(input()) #Semieje mayor de la elipse 
b=float(input()) #Semieje menor de la elipse

ae=ae(a,b)


print("\nArea de la elipse:", ae) #Impresion del resultado
print("\nArea de la elipse con dos puntos decimales: {:.2f}".format(ae))
```

**Volumen de una esfera**
$V=\frac{4}{3}\pi r^3$



```
#Volumen de una esfera.
print("Volumen de una esfera")
def V(x):
  return 4/3*np.pi*x**3 #Formula de esfera

print("Ingrese el valor del radio de la esfera")
r=float(input()) #Radio de la esfera 

V=V(r)


print("\nVolumen de la esfera:", V) 
print("\nVolumen de la esfera con dos puntos decimales: {:.2f}".format(V))
```
